# 多图布局

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
plt.rcParams['font.sans-serif'] = 'PingFang SC'
plt.rcParams['axes.unicode_minus'] = False
```

## 子图
#### `subplot()` 函数

```python
# 2 行 2 列
fig = plt.figure(figsize=(6, 4))

x = np.linspace(-np.pi, np.pi, 30)
y = np.sin(x)

# 子图 1
ax1 = plt.subplot(221)  # 2行 2列 第1个图
ax1.plot(x, y)
ax1.set_title('子图1')

# 子图 2
ax2 = plt.subplot(222)  # 2行 2列 第2个图
ax2.plot(x, y)
ax2.set_title('子图2')

# 子图 3
ax3 = plt.subplot(223)  # 2行 2列 第3个图
ax3.plot(x, y)
ax3.set_title('子图3')

# 子图 4
ax4 = plt.subplot(224)  # 2行 2列 第4个图
ax4.plot(x, y)
ax4.set_title('子图4')

# 自动调整布局
fig.tight_layout()
```

```python
# 2 行 3 图
fig = plt.figure(figsize=(6, 4))

x = np.linspace(-np.pi, np.pi, 30)
y = np.sin(x)

# 子图 1
ax1 = plt.subplot(2, 2, 1)  # 2行 2列 第1个图
ax1.plot(x, y)
ax1.set_title('子图1')

# 子图 2
ax2 = plt.subplot(2, 2, 2)  # 2行 2列 第2个图
ax2.plot(x, y)
ax2.set_title('子图2')

# 子图 3
ax3 = plt.subplot(2, 1, 2)  # 2行 1列 第2行
ax3.plot(x, y ** 2)
ax3.set_title('子图3')

# 自动调整布局
fig.tight_layout()
```

#### `subplots()` 函数

```python
x = np.linspace(0, 2 * np.pi)

# 3 行 3 列
fig, ax = plt.subplots(3, 3)
ax1, ax2, ax3 = ax
ax11, ax12, ax13 = ax1
ax21, ax22, ax23 = ax2
ax31, ax32, ax33 = ax3

# fig 设置画布大小
fig.set_figwidth(8)
fig.set_figheight(5)

# 第 1 行
ax11.plot(x, np.sin(x))
ax12.plot(x, np.cos(x))
ax13.plot(x, np.tan(x))

# 第 2 行
ax21.plot(x, np.sinh(x))
ax22.plot(x, np.cosh(x))
ax23.plot(x, np.tanh(x))

# 第 3 行
ax31.plot(x, np.sin(x)  * np.cos(x))
ax32.plot(x, np.cos(x) ** 2)
ax33.plot(x, np.tan(x) * np.sin(x))

# 自动调整布局
fig.tight_layout()
```

## 图形嵌套
#### `add_subplot()` 函数

```python
fig = plt.figure(figsize=(8, 5))  # 画布大小

# 父图
axes1 = fig.add_subplot(1, 1, 1)  # 1行 1列 第1个
axes1.plot([1, 3])

# 子图
axes1 = fig.add_subplot(2, 2, 1, facecolor='pink')  # 在父图的画布上的 2行 2列 第1个
axes1.plot([1, 3])
```

#### `axes()` 函数

```python
fig = plt.figure(figsize=(8, 5))

# 父图
x = np.linspace(0, 2 * np.pi, 30)
y = np.sin(x)
plt.plot(x, y)

# 子图 1
axes1 = plt.axes([0.55, 0.55, 0.3, 0.3])  # [left, bottom, width, heigth]
axes1.plot(x, y, color='g')

# 子图 2
axes1 = plt.axes([0.18, 0.18, 0.3, 0.3])  # [left, bottom, width, heigth]
axes1.plot(x, y, color='r')
```

## 双轴显示（组合图）

```python
fig = plt.figure(figsize=(6, 4))

x = np.linspace(0, 10 ,30)

# 图 1
axes1 = plt.gca()  # 得到当前轴域
axes1.plot(x, np.exp(x), color='r')
axes1.set_xlabel('time')  # 设置x轴标签
axes1.set_ylabel('exp', color='r')  # 设置图1的y轴标签
axes1.tick_params(axis='y', labelcolor='r')  # 设置刻度样式（指定轴，设置刻度标签颜色）

# 图 2
axes2 = axes1.twinx()  # 和图1共享x轴
axes2.plot(x, np.sin(x), color='b')
axes2.set_ylabel('sin', color='b')  # 设置图2的y轴标签
axes2.tick_params(axis='y', labelcolor='b')  # 设置刻度样式（指定轴，设置刻度标签颜色）
```
