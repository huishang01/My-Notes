# 基本图形绘制

### 导包

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### 运行时配置参数

```python
# 如果浏览器不显示图标，则加上这行代码
%matplotlib inline

# 设置中文字体，让图片中可以显示中文
plt.rcParams['font.sans-serif'] = 'PingFang SC'
# 让图片中可以显示负号(-)
plt.rcParams['axes.unicode_minus'] = False

# 如果需要用到svg矢量图，则加上这行代码
%config InlineBackend.figure_format = 'svg'
```

### 基本绘图
- `plt.plot()`

```python
# 抛物线
x = np.linspace(-5, 5, 50)  # 等差数列，范围 5 到 -5，数量 50
y = x ** 2

# 画图：线型图、折线图
plt.plot(x, y)
```

### 样式和颜色
- 样式：`'-', '--', '-.', ':', '', 'o', '^', 'v', '<', '>', 'S', '+', 'x', 'D', 'd', '1', '2', '3', '4', 'h', 'H', 'p', '|', '_'`
- 颜色：`r(红), g(绿), b(蓝), c(青), m(品红), y(黄), k(黑), w(白)`

```python
x = np.linspace(-5, 5, 50)
y = x ** 2

# color：线条颜色  ls：line style 线条样式
# plt.plot(x, y, color='red', ls='--')
plt.plot(x, y, 'b--')
```

### 画布配置
- `plt.figure()`

```python
# figsize(width, height)：画布大小
# dpi：分辨率
# facecolor：背景颜色
plt.figure(figsize=(5, 3), dpi=300, facecolor='#F8ADC9')

x = np.linspace(0, 2 * np.pi)
y = np.sin(x)
plt.plot(x, y)

# 设置网格
plt.grid()
```

### 在一个画布上绘制多个图

```python
plt.figure(figsize=(5, 3))
x = np.linspace(0, 8)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x), 'r')
plt.plot(x, -np.sin(x), 'g--')
```

### 显示图片
- `plt.show()`

```python
plt.figure(figsize=(5, 3))
x = np.linspace(0, 8)

plt.plot(x, np.sin(x))
plt.show()  # show() 写在中间会变成绘制两幅图，后面的是一个新的画布

plt.plot(x, np.cos(x), 'r')
plt.plot(x, -np.sin(x), 'g--')
plt.show()  # 立刻显示图片
```
