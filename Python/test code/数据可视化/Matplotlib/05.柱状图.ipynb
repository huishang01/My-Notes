# 柱状图

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
plt.rcParams['font.sans-serif'] = 'PingFang SC'
plt.rcParams['axes.unicode_minus'] = False
```

### 基础柱状图

```python
x = np.arange(2014, 2021)  # 生成2014-2021的数列
y = np.random.randint(1500000, 4000000, size=7)  # 在1500000-4000000之间随机生成7个数字

plt.figure(figsize=(8, 3))  # 创建画布

plt.bar(x, y, width=0.5)  # 柱子的宽度

plt.title('年度销售额', va='bottom')  # 设置柱状图标题
plt.xlabel('年份')  # 设置x轴的标签
plt.ylabel('销售额')  # 设置y轴的标签
plt.ylim(0, 4500000)  # 设置y轴范围
plt.yticks(np.arange(0, 4000001, 500000))  # 设置y轴的刻度 范围0-4000000 间隔500000

# 在柱子上显示数值（文本）
for a, b in zip(x, y):
    plt.text(
        x=a,
        y=b,
        s='{:.1f}万'.format(b / 10000),
        fontsize=9,
        ha='center',
        va='bottom',
    )
```

### 簇状柱形图

```python
x = np.arange(2014, 2021)  # 生成2014-2021的数列

plt.figure(figsize=(8, 3))  # 创建画布

w = 0.2  # 柱子的宽度
plt.bar(x+w, np.random.randint(1500000, 4000000, size=7), width=w, label='南部')  # 在1500000-4000000之间随机生成7个数字
plt.bar(x, np.random.randint(1500000, 4000000, size=7), width=w, label='中部')
plt.bar(x-w, np.random.randint(1500000, 4000000, size=7), width=w, label='北部')

plt.legend()  # 显示图例
```

### 堆叠柱状图

```python
x = np.arange(2014, 2021)  # 生成2014-2021的数列

plt.figure(figsize=(8, 3))  # 创建画布

y1 = np.random.randint(1500000, 4000000, size=7)  # 在1500000-4000000之间随机生成7个数字
y2 = np.random.randint(1500000, 4000000, size=7)
y3 = np.random.randint(1500000, 4000000, size=7)

plt.bar(x, y1, label='南部')
plt.bar(x, y2, label='中部', bottom=y1)  # bottom：设置堆叠效果
plt.bar(x, y3, label='北部', bottom=y1+y2)

plt.legend()  # 显示图例
```
