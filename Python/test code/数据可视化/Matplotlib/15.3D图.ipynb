# 3D 图

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
plt.rcParams['font.sans-serif'] = 'PingFang SC'
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 导入 3D 引擎
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

### 3D 折线图

```python
x = np.linspace(0, 100, 400)
y = np.sin(x)
z = np.cos(x)

fig = plt.figure(figsize=(5, 3))  # 设置画布
axes = Axes3D(fig, auto_add_to_figure=False)
fig.add_axes(axes)

axes.plot(x, y, z)
```

### 3D 散点图

```python
x = np.random.rand(50)  # 随机生成50个数
y = np.random.rand(50)
z = np.random.rand(50)

fig = plt.figure(figsize=(5, 3))  # 设置画布
axes = Axes3D(fig, auto_add_to_figure=False)
fig.add_axes(axes)

axes.scatter(x, y, z, color='red', s=50)
```

### 3D 柱状图

```python
x = np.arange(1, 5)

fig = plt.figure(figsize=(5, 3))  # 设置画布
axes = Axes3D(fig, auto_add_to_figure=False)
fig.add_axes(axes)

for m in x:
    axes.bar(left=np.arange(4),  # y轴
             height=np.random.randint(10, 100, size=4),  # z轴
             zs=m,  # 在x轴的第几个
             zdir='y',  # 排列方向
             width=0.6,  # 柱子的宽度
             alpha=0.6)  # 柱子的透明度

axes.set_xlabel('X轴')
axes.set_ylabel('Y轴')
axes.set_zlabel('Z轴')
```
