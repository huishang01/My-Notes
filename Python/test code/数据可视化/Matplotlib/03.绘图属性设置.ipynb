# 绘图属性设置

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
plt.rcParams['font.sans-serif'] = 'PingFang SC'
plt.rcParams['axes.unicode_minus'] = False
```

## 绘图基本属性
### 图例
- `plt.legend()`

```python
plt.figure(figsize=(6, 4))

x = np.linspace(0, 2 * np.pi)

# 添加方式 1
plt.plot(x, np.sin(x), label='sin')
plt.plot(x, np.cos(x), label='cos')

# 图例
plt.legend()
```

```python
plt.figure(figsize=(6, 4))

x = np.linspace(0, 2 * np.pi)
plt.plot(x, np.sin(x), label='sin')
plt.plot(x, np.cos(x), label='cos')

# 图例
# 添加方式 2
plt.legend(['sin', 'cos'],
           fontsize=18,  # 字号
           loc='center',  # 位置 默认为 best 放在最佳位置
           # loc=9,  # 也可以写数字
           ncol=2,  # 显示列数 默认1列
           bbox_to_anchor=[0, 1, 1, 0.2])  # 具体位置 [x, y, width, heigth]
```

### 线条属性
- `data`：数据 列表
- `color`：线条颜色
- `ls`：线条类型
- `lw`：线条宽度
- `marker`：标记类型
- `markersize`：标记大小
- `markeredgecolor`：标记颜色
- `markeredgewidth`：标记描边宽度
- `mfc`：标记背景颜色
- `label`：标签
- `alpha`：透明度

```python
fig = plt.figure(figsize=(6, 4))
x = np.linspace(0, 2 * np.pi, 20)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1 * y2,
         color='orange',  # 线条颜色
         ls=':',  # 线条类型
         lw=2,  # 线条宽度
         marker='*',  # 标记类型
         markersize='10',  # 标记大小
         markeredgecolor='m',  # 标记颜色
         markeredgewidth=2,  # 标记描边宽度
         mfc='r',  # 标记背景颜色 marker face color
         label='sinx+cosx',  # 标签
         alpha=0.5)  # 透明度

plt.legend()
```

### 坐标轴刻度
- `plt.xticks()`
- `plt.yticks()`

```python
plt.figure(figsize=(6, 3))

x = np.linspace(0, 10)
y = np.sin(x)

plt.xticks(ticks=np.arange(0, 11, 1), fontsize=15, color='r')  # 设置x轴的刻度 列表

plt.yticks(ticks=[-1, 0, 1],  #刻度值
           labels=['min', 0, 'max'],  # 刻度标签
           fontsize=15,  # 刻度标签字号
           color='b',  # 刻度标签颜色
           ha='right')  # 刻度标签位置
          
plt.plot(x, y)
```

### 坐标轴范围
- `plt.xlim()`
- `plt.ylim()`

```python
plt.figure(figsize=(5, 3))

x = np.linspace(0, 2 * np.pi)
y = np.sin(x)

plt.xlim(-2, 8)  # 设置x轴范围
plt.ylim(-2, 2)  # 设置y轴范围

plt.plot(x, y, c='r')
```

### 坐标轴配置
- `plt.axis()`

```python
plt.figure(figsize=(5, 3))

x = np.linspace(0, 2 * np.pi)
y = np.sin(x)

plt.axis([-2, 8, -2, 2])  # 设置刻度范围 [xmin, xmax, ymin, ymax]

# option 坐标轴配置
# off：不显示坐标轴
# equal：让x轴和x轴的刻度距离相等
# scaled：自动缩放
# tight：紧凑型自动缩放
# square：让画布变成正方形
plt.axis('scaled')

plt.plot(x, y, c='r')
```

### 标题 和 网格
- `plt.title()`
- `plt.grid()`

```python
fig = plt.figure(figsize=(5, 3))

x = np.linspace(0, 10)
y = np.sin(x)

# 标题
plt.title('sin曲线',  # 标题
          loc='center',  # 位置
          y=1.1,  # y轴位置
          fontsize=16)  # 字号

plt.suptitle('副标题',  # 副标题
             y=0.95,  # y轴位置
             x=0.51,  # x轴位置
             fontsize=12)  # 字号

# 网格线
plt.grid(ls=':',  # 线条样式
         lw=0.5,  # 线条宽度
         color='grey',  # 线条颜色
         axis='y')  # 显示y轴的线  both：x轴y轴都显示

plt.plot(x, y)
```

### 标签
- `plt.xlabel()`
- `plt.ylabel()`

```python
fig = plt.figure(figsize=(5, 3))

x = np.linspace(0, 10)
y = np.sin(x)

plt.xlabel('y=sin(x)', fontsize=16, rotation=0, ha='center')

plt.ylabel('y=sin(x)',  # 标签名
           fontsize=16,  # 标签字号
           rotation=0,  # 旋转角度
           ha='right',  # 水平对齐方式
           x=0,  # 标签的x轴坐标
           y=0.5,)  # 标签的y轴坐标

plt.title('正弦曲线')

plt.plot(x, y)
```

### 文本
- `plt.text()`

```python
fig = plt.figure(figsize=(5, 3))

x = np.linspace(1, 10, 10)
y = np.tan(x) ** 2

plt.plot(x, y, ls='--', marker='o')  # 设置线条样式

for a, b in zip(x, y):
    plt.text(
        x=a,  # x轴坐标
        y=b+4,  # y轴坐标
        s=int(b),  # 文本内容
        fontsize=12,  # 文本字号
        color='b',  # 文本颜色
        ha='center',  # 水平对齐方式
        va='center',  # 上下对齐方式
    )

plt.ylim(-10, 60)  # 设置y轴范围
```

### 标注
- `plt.annotate()`

```python
fig = plt.figure(figsize=(5, 3))

x = np.linspace(1, 10, 10)
y = np.tan(x) ** 2

plt.plot(x, y, ls='--', marker='o')  # 设置线条样式

plt.annotate(
    text='最高',  # 标注内容
    xy=(7.8, 46),  # 标注点的位置，箭头的指向点
    xytext=(6, 40),  # 标注文本的位置
    arrowprops={  # 箭头样式
        'width': 3,  # 箭头线的宽度
        'headwidth': 7,  # 箭头头部的宽度
        'shrink': 0.1,  # 箭头的长度 0-1
        'headlength': 7,  # 箭头头部长度
        'facecolor': 'm',  # 箭头的背景颜色
        'edgecolor': 'b',  # 箭头描边的颜色
    }
)
```

### 保存图片
- `plt.savefig()`

```python
fig = plt.figure(figsize=(5, 3))

x = np.linspace(0, 2 * np.pi)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

# plt.savefig('sincos.png')

fig.savefig(
    fname='sincos.png',  # 保存路径
    dpi=200,  # 像素密度
    facecolor='grey',  # 背景颜色
    pad_inches=10,  # 内边距
)
```
