# 图像处理

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### 读取图片
- `plt.imread`

```python
img = plt.imread('data/01.jpg')
img  # 彩色图：RGB
```

```python
img.shape  # 图片大小为537 * 474
```

### 显示图片
- `plt.imshow()`

```python
plt.imshow(img)
```

### 垂直翻转

```python
# plt.imshow(img, origin='lower')
plt.imshow(img[::-1])
```

### 水平翻转

```python
plt.imshow(img[:, ::-1])
```

### 截图

```python
plt.imshow(img[200:400, 200:])
```

### 保存图片
- `plt.imsave()`

```python
img2 = img[200:400, 200:]
plt.imsave('data/img2.jpg', img2)
```
