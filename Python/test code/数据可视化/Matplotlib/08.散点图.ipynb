# 散点图
散点图用于在水平轴和垂直轴上绘制数据点，常用于描述两点之间变量的关系

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
plt.rcParams['font.sans-serif'] = 'PingFang SC'
plt.rcParams['axes.unicode_minus'] = False
```

### 基础散点图

```python
x = range(1, 7, 1)
y = range(10, 70, 10)

plt.scatter(x, y)
```

### 气泡图

```python
data = np.random.randn(100, 2)  # 100行 2列的随机数的数组

plt.figure(figsize=(5, 3))  # 设置画布大小
s = np.random.randint(50, 200, size=100)  # 生成100个范围为50-100的随机数的数组
c = np.random.randn(100)  # 生成100个随机数的数组

plt.scatter(data[:, 0], data[:, 1], s=s, c=c, alpha=0.6)  # 所有行，第0列和第1列
```

### 散点图2

```python
array1 = np.random.randint(4000, 40000, size=30)
x = np.sort(array1)
array2 = np.random.randint(5000, 100000, size=30)
y = np.sort(array2)

plt.figure(figsize=(5, 3))  # 设置画布大小
plt.scatter(x, y)

plt.title('广告费用和销售收入之间的关系')
plt.xlabel('广告费用')
plt.ylabel('销售收入')
```

### 六边形图

```python
array1 = np.random.randint(4000, 40000, size=30)
x = np.sort(array1)
array2 = np.random.randint(5000, 100000, size=30)
y = np.sort(array2)

plt.figure(figsize=(5, 3))  # 设置画布大小
plt.hexbin(x, y,
           gridsize=20,  # 点的大小
           cmap='rainbow')  # 颜色设置为彩虹色

plt.title('广告费用和销售收入之间的关系')
plt.xlabel('广告费用')
plt.ylabel('销售收入')
```
