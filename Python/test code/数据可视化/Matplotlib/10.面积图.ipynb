# 面积图
面积图又称区域图，和折线图差不多，强调 Y 轴随 X 轴而变化的程度

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
plt.rcParams['font.sans-serif'] = 'PingFang SC'
plt.rcParams['axes.unicode_minus'] = False
```

### 基础面积图

```python
x = [1, 2, 3, 4, 5]
y = np.random.randint(10, 100, 5)

plt.stackplot(x, y)
plt.plot(x, y)
```

### 面积图2

```python
x = np.arange(2014, 2021)
y = np.random.randint(165000, 2850000, 7)

plt.stackplot(x, y)
plt.plot(x, y)
plt.show()
```
