# 直方图
统计出现的次数

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
plt.rcParams['font.sans-serif'] = 'PingFang SC'
plt.rcParams['axes.unicode_minus'] = False
```

### 基础直方图

```python
x = np.random.randint(0, 10, 100)
x
```

```python
# 统计每个数出现的次数
pd.Series(x).value_counts()
```

```python
plt.figure(figsize=(5, 3))

# plt.hist(x)  # 统计出现的次数

# plt.hist(x, bins=5)  # bins：组数
plt.hist(x, bins=[0, 3, 5, 8, 9])  # bins还可以自定义分组

plt.xticks(range(10))

plt.show()
```

### 直方图的常用属性

```python
x = np.random.randint(50, 121, size=1000)  # 在50-100之间随机生成1000个数字

plt.hist(x,  # 添加数据
         density=True,  # 使用概率分布
         bins=range(50, 121, 5),  # 分组 范围50-120 间隔5 的数列
         facecolor='b',  # 背景颜色
         alpha=0.5,  # 透明度
         edgecolor='w')  # 边框颜色

plt.show()
```
