# 雷达图
雷达图是从原点开始的、轴上表示三个或多个定量变量的、二维图标的形式、显示多变量数据的图形方法

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
plt.rcParams['font.sans-serif'] = 'PingFang SC'
plt.rcParams['axes.unicode_minus'] = False
```

### 基础雷达图

```python
x = np.linspace(0, 2 * np.pi, 6, endpoint=False)  # 从0到2*pi生成8个等差数列，不包尾
y = np.random.randint(10, 100, 6)  # 生成6个范围为10到100的随机数数组

x = np.concatenate((x, [x[0]]))  # 让数组添加一个值，这个值与数组第一个数一样
y = np.concatenate((y, [y[0]]))

plt.figure(figsize=(5, 5))  # 画布大小
axes = plt.subplot(111, polar=True)
axes.plot(x, y, 'o--')  # 点线样式
axes.fill(x, y, alpha=0.3)  # 填充
axes.set_rgrids([20, 40, 60, 80, 100], fontsize=10)  # 刻度
plt.show()
```
