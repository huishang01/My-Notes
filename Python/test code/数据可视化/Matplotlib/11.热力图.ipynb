# 热力图
热力图是一种通过对色块着色来显示数据的统计数据表。绘图时需要指定颜色映射的规则

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
plt.rcParams['font.sans-serif'] = 'PingFang SC'
plt.rcParams['axes.unicode_minus'] = False
```

### 基础热力图

```python
y = ['广东', '广西', '湖南', '湖北', '江西', '四川', '福建', '江苏', '河南', '河北', '山东', '山西']
x = ['产品A', '产品B', '产品C', '产品D', '产品E', '产品F', '产品G']
data = np.random.randint(1300, 9900, (12, 7))

plt.figure(figsize=(8, 8))  # 设置画布

plt.imshow(data, cmap='Blues')  # 绘制热力图

plt.xticks(range(len(x)), x)  # x轴刻度
plt.yticks(range(len(y)), y)  # y轴刻度

# 添加文字
for i in range(len(x)):  # 遍历x轴得到每列
    for j in range(len(y)):  # 遍历y轴得到每一列中的每一行
        plt.text(x=i,y=j,  # 坐标
                 s=data[j, i],  # 数据
                 ha='center',  # 水平对齐
                 va='center')  # 垂直对齐

plt.colorbar()  # 图例
plt.show()
```
