# 极坐标图
极坐标系是一个二维坐标系统，该坐标系统中的任意位置可由一个夹角和一段相对原点(极点)的距离来表示

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
plt.rcParams['font.sans-serif'] = 'PingFang SC'
plt.rcParams['axes.unicode_minus'] = False
```

### 基础极坐标图

```python
n = 8
x = np.linspace(0, 2 * np.pi, n, endpoint=False)  # 从0到2*pi生成8个等差数列，不包尾
height = np.random.randint(3, 15, size=n)  # 生成8个范围为3到15的随机数数组
width = 2 * np.pi / 8  # 每个扇区的宽度，将图分为八份
color = np.random.rand(8, 3)  # 随机颜色

axes = plt.subplot(111, projection='polar')
axes.bar(x=x,  # 数据
         height=height,  # 扇区高度
         width=width,  # 扇区宽度
         bottom=0,  # 数值开始位置
         color=color)  # 扇区颜色
```
