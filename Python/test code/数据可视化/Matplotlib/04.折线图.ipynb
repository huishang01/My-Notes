# 折线图

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
plt.rcParams['font.sans-serif'] = 'PingFang SC'
plt.rcParams['axes.unicode_minus'] = False
```

### 绘制一条线

```python
plt.figure(figsize=(5, 3))

x = ['Mon', 'Tues', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']
y = [20, 30, 60, 30, 50, 70, 40]

# 线条样式
plt.plot(x, y, c='g', marker='D', markersize=5)

# 标签和标题
plt.xlabel('星期')
plt.ylabel('活跃度')
plt.title('不知道什么东西的活跃度')

# x轴刻度范围
plt.ylim(10, 80)

# 添加文本
for a, b in zip(x, y):
    plt.text(a, b+2, b, ha='center', va='bottom')

plt.show()
```

### 绘制多条线

```python
plt.figure(figsize=(10, 3))  # 创建画布

x = np.arange(1, 13)  # 生成1-12的数列
y1 = np.random.randint(60, 100, size=12)  # 在60-100之间随机生成12个数字
y2 = np.random.randint(60, 100, size=12)
y3 = np.random.randint(60, 100, size=12)

plt.plot(x, y1, label='语文', ls='--', marker='p', mfc='w', markersize=5)  # 设置线条和标记的样式
plt.plot(x, y2, label='数学', ls='-.', marker='o', mfc='w', markersize=5)
plt.plot(x, y3, label='英语', ls=':', marker='D', mfc='w', markersize=4)

plt.xticks(x)  # 设置x轴的刻度
plt.yticks(np.arange(0, 101, 10))  # 设置y轴的刻度 范围10-100 间隔10

plt.ylim(0, 110)  # 设置y轴的范围

plt.grid(ls='--', lw=0.5, axis='y')  # 设置网格线

plt.xlabel('月份')  # 设置x轴的标签
plt.ylabel('分数')  # 设置y轴的标签
plt.title('月成绩变化趋势')  # 设置标题

plt.legend()  # 显示图例

plt.show()
```
