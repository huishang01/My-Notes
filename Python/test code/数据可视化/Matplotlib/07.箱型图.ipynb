# 箱型图
也被称为盒须图，它能显示出一组数据的最大值、最小值、中位数以及上下四分位数
- 最大值
- 3/4  Q3
- 2/4  中位数
- 1/4  Q1
- 最小值
- 
- 异常值

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
plt.rcParams['font.sans-serif'] = 'PingFang SC'
plt.rcParams['axes.unicode_minus'] = False
```

### 基础箱型图

```python
x = [1, 3, 6, 8, 4, 9, 4, 3, 7, -5]

plt.figure(figsize=(4, 4))
plt.boxplot(x)
plt.show()
```

### 绘制多个箱型图

```python
x1 = np.random.randint(10, 100, 100)
x2 = np.random.randint(10, 100, 100)
x3 = np.random.randint(10, 100, 100)

plt.figure(figsize=(5, 4))
plt.boxplot([x1, x2, x3])  # 传入列表
plt.show()
```

```python
data = np.random.normal(size=(500, 4))  # 500行 4列 的二维数组
labels = ['A', 'B', 'C', 'D']

plt.boxplot(data,  # 添加数据
            labels=labels,  # 添加标签
            notch=True,  # 箱型图样式
            sym='gv'  # 异常值样式 颜色和marker
           )
plt.show()
```
