### 基础柱状图

```python
from pyecharts.charts import Bar

# 在 jupyter 里绘制需要导包
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
```

```python
bar = Bar()
bar.add_xaxis(['衬衣', '羊毛衫', '西装', '裤子', '鞋子', '袜子'])
bar.add_yaxis('商家', [50, 90, 120, 30, 70, 10])
```

生成 HTML 文件

```python
bar.render('Charts/01_bar.html')
```

在 Notebook 里绘制

```python
bar.load_javascript()
```

```python
bar.render_notebook()
```

链式调用

```python
bar = (
    Bar()
    .add_xaxis(['衬衣', '羊毛衫', '西装', '裤子', '鞋子', '袜子'])
    .add_yaxis('商家', [50, 90, 120, 30, 70, 10])
)
```

```python
bar.render_notebook()
```

### Options 配置项

```python
from pyecharts import options as opts
```

```python
bar = (
    Bar()
    .add_xaxis(['衬衣', '羊毛衫', '西装', '裤子', '鞋子', '袜子'])
    .add_yaxis('商家', [50, 90, 120, 30, 70, 10])
    .set_global_opts(title_opts=opts.TitleOpts(title='主标题', subtitle='副标题'))
)
```

```python
bar.render_notebook()
```

### 使用主题

```python
from pyecharts.globals import ThemeType
```

```python
bar = (
    Bar(
        init_opts=opts.InitOpts(theme=ThemeType.DARK)
    )
    .add_xaxis(['衬衣', '羊毛衫', '西装', '裤子', '鞋子', '袜子'])
    .add_yaxis('商家A', [50, 90, 120, 30, 70, 10])
    .add_yaxis('商家B', [40, 80, 110, 20, 60, 8])
)
```

```python
bar.render_notebook()
```
